### Tratamiento y selección por cada variable

```python
# creamos una copia de la bd
df1 = df.copy() 
df1.describe()
```

```python
# Obtenermos la distribución de los datos sobre la variable X2
# X2: Gender (1 = male; 2 = female)
print ("Distribución variable X2: \n",df1['X2'].value_counts())

# X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

print ("Distribución variable X3: \n",df1['X3'].value_counts())

# X4: Marital status (1 = married; 2 = single; 3 = others).
print ("Distribución variable X4: \n",df1['X4'].value_counts())
```

realizamos tratamiento de las variables:

- X2: Gender (1 = male; 2 = female). 
- X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 
- X4: Marital status (1 = married; 2 = single; 3 = others). 
- X5: Age (year). 
- X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

Las variables tienen entre 1 y 14 datos null, procederemos a eliminarlos

Obtenemos el número de registros sobre estas variables

```python
v1 = ['X2','X3','X4','X5','X6','X7','X8','X9','X10','X11']
null_=df1[df1[v1].isna()]
print(null_)

df1=df1[v1].dropna()
```

```python
# Validamos que sobre las variables seleccionadas no existan valores nulos
df1.isnull().sum()
```

```python
## Para las variables X12 - X17, utilizaremos la media para reemplazar los valore nulos

df2 = df.copy() 
df2.head

# Reemplazamos los valor null con la media 

df2.X12.fillna(df2.X12.mean(), inplace=True)
df2.X13.fillna(df2.X13.mean(), inplace=True)
df2.X14.fillna(df2.X14.mean(), inplace=True)
df2.X15.fillna(df2.X15.mean(), inplace=True)
df2.X16.fillna(df2.X16.mean(), inplace=True)
df2.X17.fillna(df2.X17.mean(), inplace=True)

df2.isnull().sum()
```

```python
# Obtenemos la moda sobre las variables X18 - X23

var_moda=['X18','X19','X20','X21','X22','X23'] # definimos variables

# calculamos la moda 
favs=df2[var_moda].mode()
favs
```

```python
## Para las variables X18 - X23, utilizaremos moda para reemplazar los valore nulos

df3 = df2.copy() 
df3.head

# Reemplazamos los valor null con la moda

df3.X18.fillna(df3.X18.mode()[0], inplace=True)
df3.X19.fillna(df3.X19.mode()[0], inplace=True)
df3.X20.fillna(df3.X20.mode()[0], inplace=True)
df3.X21.fillna(df3.X21.mode()[0], inplace=True)
df3.X22.fillna(df3.X22.mode()[0], inplace=True)
df3.X23.fillna(df3.X23.mode()[0], inplace=True)

df3.isnull().sum()
```
